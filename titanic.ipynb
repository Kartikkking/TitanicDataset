```python
import pandas as pd
import numpy as np
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('titanic')
print("First few rows of the Titanic dataset:")
print(df.head())
```

```python
print("\nBasic statistics of the Titanic dataset:")
print(df.describe(include='all'))
```

```python
# Plotting the histogram of the 'age' feature
plt.figure(figsize=(10, 6))
sns.histplot(df['age'].dropna(), bins=30, kde=True)
plt.title('Distribution of Age in Titanic Dataset')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Box plot for 'age' grouped by 'survived'
plt.figure(figsize=(10, 6))
sns.boxplot(x='survived', y='age', data=df)
plt.title('Box Plot of Age by Survival Status')
plt.xlabel('Survived')
plt.ylabel('Age')
plt.show()
```

```python
# Scatter plot for 'pclass' vs 'fare'
plt.figure(figsize=(5, 3))
sns.scatterplot(x='pclass', y='fare', hue='survived', data=df, palette='coolwarm')
plt.title('Scatter Plot of Passenger Class vs Fare by Survival Status')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.show()

```

```python
# Create scatter plots for numerical features
sns.pairplot(df[['age', 'fare', 'pclass', 'survived']], hue='survived', palette='Set1')
plt.show()
```

```python
# Create box plots for categorical features vs numerical features
plt.figure(figsize=(12, 6))
sns.boxplot(x='pclass', y='age', data=df)
plt.title('Box plot of Age vs Pclass')
plt.show()
```

```python
df['sex_numeric'] = df['sex'].map({'male': 0, 'female': 1})

# Select only numeric columns for correlation matrix
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix
corr_matrix = numeric_df.corr()

# Plot the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Heatmap of Correlation between Numerical Features in Titanic Dataset')
plt.show()
```
